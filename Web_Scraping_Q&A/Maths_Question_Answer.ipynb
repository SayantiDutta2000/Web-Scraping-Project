```python
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get('https://www.studyadda.com/question-bank/11th-class/23/mathematics/3')
soup = BeautifulSoup(source.text, 'lxml')
links = soup.find_all('a', class_='subjectbtn btn btn-primary btn-raised btn-lg')

Links = []
for l in links:
    Links.append(l.get('href'))
Links
```

```python
f = open('Class_11_Maths.csv', 'w', encoding='UTF-8')

header = "Question, Option_A, Option_B, Option_C, Option_D\n"
f.write(header)

for l in Links:
    Request = requests.get(l)
    Soup = BeautifulSoup(Request.text, 'lxml')
    
    LINKS = []
    row = Soup.find('div', class_='row recent_ques')
    for i in row.find_all('a'):
        LINKS.append(i.get('href'))

    for j in LINKS[:1]:
        REQUEST = requests.get(j)
        SOUP = BeautifulSoup(REQUEST.text, 'lxml')
        containers = SOUP.find_all('li', class_='element-item singlechoice')

        for container in containers:
            Text = container.find_all('p')
            question = Text[0].text
            question = question[18:]
            question = question.replace('}{', '/')
            question = question.replace('[', '')
            question = question.replace(']', '')
            question = question.replace('{+}', ' + ')
            question = question.replace('{-}', ' - ')
            question = question.replace('=', ' = ')
            question = question.replace('\\', '')
            question = question.replace('{', '')
            question = question.replace('}', '')
            question = question.replace('frac', '')
            question = question.replace('mathbf', '')
            option_a = Text[1].text
            option_a = option_a.replace('done', '')
            option_a = option_a.replace('clear', '')
            option_a = option_a.replace('}{', '/')
            option_a = option_a.replace('[', '')
            option_a = option_a.replace(']', '')
            option_a = option_a.replace('{+}', ' + ')
            option_a = option_a.replace('{-}', ' - ')
            option_a = option_a.replace('=', ' = ')
            option_a = option_a.replace('\\', '')
            option_a = option_a.replace('{', '')
            option_a = option_a.replace('}', '')
            option_a = option_a.replace('frac', '')
            option_a = option_a.replace('mathbf', '')
            option_a = " ".join(option_a.split())
            option_a = option_a[2:]
            option_b = Text[2].text
            option_b = option_b.replace('done', '')
            option_b = option_b.replace('clear', '')
            option_b = option_b.replace('}{', '/')
            option_b = option_b.replace('[', '')
            option_b = option_b.replace(']', '')
            option_b = option_b.replace('{+}', ' + ')
            option_b = option_b.replace('{-}', ' - ')
            option_b = option_b.replace('=', ' = ')
            option_b = option_b.replace('\\', '')
            option_b = option_b.replace('{', '')
            option_b = option_b.replace('}', '')
            option_b = option_b.replace('frac', '')
            option_b = option_b.replace('mathbf', '')
            option_b = " ".join(option_b.split())
            option_b = option_b[2:]
            option_c = Text[3].text
            option_c = option_c.replace('done', '')
            option_c = option_c.replace('clear', '')
            option_c = option_c.replace('}{', '/')
            option_c = option_c.replace('[', '')
            option_c = option_c.replace(']', '')
            option_c = option_c.replace('{+}', ' + ')
            option_c = option_c.replace('{-}', ' - ')
            option_c = option_c.replace('=', ' = ')
            option_c = option_c.replace('\\', '')
            option_c = option_c.replace('{', '')
            option_c = option_c.replace('}', '')
            option_c = option_c.replace('frac', '')
            option_c = option_c.replace('mathbf', '')
            option_c = " ".join(option_c.split())
            option_c = option_c[2:]
            option_d = Text[4].text
            option_d = option_d.replace('done', '')
            option_d = option_d.replace('clear', '')
            option_d = option_d.replace('}{', '/')
            option_d = option_d.replace('[', '')
            option_d = option_d.replace(']', '')
            option_d = option_d.replace('{+}', ' + ')
            option_d = option_d.replace('{-}', ' - ')
            option_d = option_d.replace('=', ' = ')
            option_d = option_d.replace('\\', '')
            option_d = option_d.replace('{', '')
            option_d = option_d.replace('}', '')
            option_d = option_d.replace('frac', '')
            option_d = option_d.replace('mathbf', '')
            option_d = " ".join(option_d.split())
            option_d = option_d[2:]
            f.write( question.replace(",", "|") + "," 
                    + option_a.replace(",", "|") + "," 
                    + option_b.replace(",", "|") + "," 
                    + option_c.replace(",", "|") + "," 
                    + option_d.replace(",", "|") + "," 
                    + "\n")

f.close()
```

```python
f = open('Maths_Class11_Answer.csv', 'w')

for l in Links:
    Request = requests.get(l)
    Soup = BeautifulSoup(Request.text, 'lxml')
    
    LINKS = []
    row = Soup.find('div', class_='row recent_ques')
    for i in row.find_all('a'):
        LINKS.append(i.get('href'))

    for k in LINKS[:1]:
        REQUEST = requests.get(k)
        SOUP =BeautifulSoup(REQUEST.text, 'lxml')

        containers = SOUP.find_all('li', class_='element-item singlechoice')

        for container in containers:
            solution = container.find('a', target = '_blank')['href']
            request1 = requests.get(solution)
            soup1 =BeautifulSoup(request1.text, 'lxml')
            correct_answer = soup1.find('p', class_='ans_panel').text
            correct_answer = correct_answer.replace('Correct Answer:', '')
            correct_answer = " ".join(correct_answer.split())
            f.write(correct_answer + "\n")
f.close()
```
